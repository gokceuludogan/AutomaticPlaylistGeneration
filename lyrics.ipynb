```python
import pandas as pd
from sklearn.decomposition import LatentDirichletAllocation
```

```python
songs_df = pd.read_csv('msd_genre_dataset.csv')
```

```python
f = open('mxm_dataset_train.txt', 'r')
lines = [x for x in f.readlines() if not x.startswith('#')]
top_words = lines[0][1:].split(',')
print(top_words[:4])
splitted_lines = [x.split(',') for x in lines[1:]]
song_word_counts = {x[0]:{top_words[int(y.split(':')[0])-1]:int(y.split(':')[1]) for y in x[2:]} for x in splitted_lines[1:]}
song_names = [x[0] for x in splitted_lines]
```

```python
song_name_df = pd.DataFrame(song_names, columns=['track_id'])
word_count_df = pd.DataFrame(song_word_counts)
```

```python
word_count_df.fillna(0, inplace=True)
```

```python
word_count_df = word_count_df.transpose()
word_count_df
```

```python
word_count_df.index.name = 'track_id'
word_count_df
```

```python
word_count_df.reset_index(inplace = True)
```

```python
result = pd.merge(songs_df, song_name_df, on='track_id')
result
```

```python
merged_words_df = pd.merge(result[['track_id']], word_count_df, on='track_id')
```

```python
merged_words_df
```

```python
merged_words_df.set_index('track_id', inplace = True)
```

```python
merged_words_df
```

```python
lda = LatentDirichletAllocation()
new_word_feature = lda.fit_transform(merged_words_df)
```

```python
new_word_feature.shape
```

```python
lda_feature_df = pd.DataFrame(new_word_feature)
lda_feature_df
```

```python
df = pd.merge(result, lda_feature_df, left_index=True, right_index=True)
df
```
